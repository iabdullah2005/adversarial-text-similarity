```python
import json
import random
import pandas as pd

file_path = "../data/raw/arxiv-metadata-oai-snapshot.json"
sample_size = 10000  # start with 10k

sampled_abstracts = []

with open(file_path, 'r') as f:
    for i, line in enumerate(f):
        if len(sampled_abstracts) >= sample_size:
            break
        
        record = json.loads(line)
        abstract = record.get("abstract", "")
        
        if abstract and 50 < len(abstract.split()) < 500:
            sampled_abstracts.append({
                "id": record.get("id"),
                "abstract": abstract
            })

df = pd.DataFrame(sampled_abstracts)
df.head()
```

```python
df.to_csv("../data/processed/abstracts_clean.csv", index=False)
```
